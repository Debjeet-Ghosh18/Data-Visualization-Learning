```python
import seaborn as sns
import pandas as pd 
import cufflinks as cf 
from plotly.offline import iplot
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips['total_bill'].plot()
```

```python
tips.groupby('day')['tip'].mean().iplot(kind='bar')
```

```python
tips.plot(kind='scatter', x='total_bill', y='tip')
```

```python
import cufflinks as cf
cf.go_offline()

tips[['total_bill', 'day']].iplot(
    kind='box',
    x='day',
    y='total_bill'
)
```

```python
tips.groupby('day')['tip'].mean().iplot(kind='bar')
```

```python
tips.iplot(kind='scatter', x= 'total_bill',y='tip',mode = 'markers')
```

```python
import plotly.express as px

px.scatter(
    tips,
    x="total_bill",
    y="tip",
    color="sex",
    size="size",
    hover_data=["day", "time"]
)
```

```python
px.box(
    tips,
    x="smoker",
    y="tip",
    color="sex"
)
```

```python
px.histogram(
    tips,
    x="tip",
    color="sex",
    nbins=30,
    barmode="overlay"
)
```

```python
px.bar(
    tips,
    x="day",
    y="tip",
    color="day",
    barmode="group",
    title="Average Tip by Day",
)
```

```python
px.violin(
    tips,
    x="day",
    y="tip",
    color="sex",
    box=True,
    points="all"
)
```

```python
px.pie(
    tips,
    names="time",
    title="Orders by Time"
)
```

```python
px.scatter(
    tips,
    x="total_bill",
    y="tip",
    facet_col="day",
    color="sex"
)
```

```python
px.imshow(
    tips[["total_bill", "tip", "size"]].corr(),
    text_auto=True,
    title="Correlation Heatmap"
)
```

```python
px.treemap(
    tips,
    path=["day", "time"],
    values="tip",
    title="Tips by Day and Time"
)
```

```python
px.sunburst(
    tips,
    path=["day", "time", "sex"],
    values="tip",
    title="Tip Distribution"
)
```

```python
px.parallel_coordinates(
    tips,
    dimensions=["total_bill", "tip", "size"],
    color="tip"
)
```

```python
px.area(
    tips,
    x="day",
    y="tip",
    color="sex",
    title="Tips by Day"
)
```
