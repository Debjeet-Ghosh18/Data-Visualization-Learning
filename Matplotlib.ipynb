```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

**Matplotlib Basics**

```python
x = np.linspace(0,5,11)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

**Subplot in Matplotlib**

```python
plt.suptitle("Subplots in Matplotlib")
plt.subplot(2,2,1)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(a**2,b**2)
plt.subplot(2,2,3)
plt.plot(x,x)
plt.subplot(2,2,4)
plt.plot(x,y)
```

**Object Oriented Way**

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch)

```python
fig = plt.figure(figsize = (5.0,5.0),dpi=100)
axis1 = fig.add_axes([.1,.2,.9,.9])
axis1.set_title("Example1")
axis1.plot(x,y)
axis2 = fig.add_axes([.2,.3,.6,.6])
axis2.set_title("Example2")
axis2.plot(x,y)
```

### Saving Plot

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(x, y)
ax.set_title("Saving the image")
plt.savefig("BasicPlot.png")
```
